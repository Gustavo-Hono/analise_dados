```python
import pandas as pd

caminho = '/content/drive/MyDrive/cars/CarsData.csv'
df = pd.read_csv(caminho)
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1) Quantidade de carros por ano
cars_per_year = (
    df.dropna(subset=['year'])
      .groupby('year')
      .size()
      .sort_index()
)

plt.figure(figsize=(10,5))
cars_per_year.plot(kind='bar')
plt.title('Quantidade de registros por ano do veículo')
plt.xlabel('Ano')
plt.ylabel('Contagem')
plt.tight_layout()
plt.show()

if len(cars_per_year) >= 2:
    first_year, last_year = cars_per_year.index[0], cars_per_year.index[-1]
    t0, tn = int(cars_per_year.iloc[0]), int(cars_per_year.iloc[-1])
    trend = "aumentou" if tn > t0 else ("diminuiu" if tn < t0 else "permaneceu estável")
    print(f"[Insight] A presença no dataset {trend} de {t0} (em {first_year}) para {tn} (em {last_year}).")
```

```python
bins   = [0, 10_000, 20_000, 30_000, 40_000, 60_000, 80_000, np.inf]
labels = ["0–10k","10–20k","20–30k","30–40k","40–60k","60–80k","80k+"]
target_bands = ["20–30k", "30–40k"]
min_n = 100
D = df[df["year"].between(2015, 2019)].copy()

def keep_min_count(s, min_n=100):
    """Retorna os valores de s que têm pelo menos min_n ocorrências."""
    vc = s.value_counts(dropna=True)
    return set(vc[vc >= min_n].index)

# Banda de km (para “mix justo”)
D["km_band"] = pd.cut(D["mileage"], bins=bins, labels=labels, right=False)

# GRÁFICO 1 — Preço mediano por ano (linhas)
# Objetivo: comparar a evolução de preço entre 2015–2019
#           controlando por combustível e por FAIXA DE KM (20–40k).
# Leitura: cada linha é um combustível.
Dd = D[D["km_band"].isin(target_bands)].copy()
need = ["fuelType","year","price","km_band"]


agg_med = (
    Dd.dropna(subset=["fuelType","year","price"])
      .groupby(["fuelType","year"])["price"].median()
      .rename("price_median")
      .reset_index()
)
agg_cnt = (
    Dd.dropna(subset=["fuelType","year","price"])
      .groupby(["fuelType","year"]).size()
      .rename("n")
      .reset_index()
)
G = agg_med.merge(agg_cnt, on=["fuelType","year"])
G = G[G["n"] >= min_n]

# Mantém somente combustíveis que aparecem em TODOS os anos do recorte
years_full = [y for y in sorted(D["year"].dropna().unique()) if 2015 <= y <= 2019]
valid_fuels = [ft for ft, g in G.groupby("fuelType") if set(g["year"]) >= set(years_full)]
G = G[G["fuelType"].isin(valid_fuels)]

plt.figure(figsize=(9,5))
for ft, sub in G.sort_values(["fuelType","year"]).groupby("fuelType"):
    plt.plot(sub["year"], sub["price_median"], marker='o', label=f"{ft} (n≥{min_n}/ano)")
plt.title(f"Preço mediano por ano (2015–2019)\nControlado por combustível e km: {', '.join(target_bands)}")
plt.xlabel("Ano"); plt.ylabel("Preço mediano")
plt.legend(title="Combustível", frameon=False)
plt.tight_layout()
plt.show()

# GRÁFICO 2 — MPG mediano por combustível (barras horizontais)
# Objetivo: comparar eficiência típica por tipo de combustível.
# Leitura: valores maiores = consome menos; “Other” costuma ser mix heterogêneo.
if {'fuelType','mpg'}.issubset(D.columns):
    fuels_big = keep_min_count(D['fuelType'], 100)
    mpg_median_by_fuel = (
        D[D['fuelType'].isin(fuels_big)]
         .dropna(subset=['fuelType','mpg'])
         .groupby('fuelType')['mpg']
         .median()
         .sort_values()
    )
    plt.figure(figsize=(9,6))
    mpg_median_by_fuel.plot(kind='barh')
    plt.title('MPG mediano por tipo de combustível (2015–2019, ≥100 amostras)')
    plt.xlabel('MPG mediano'); plt.ylabel('Combustível')
    plt.tight_layout()
    plt.show()

```
